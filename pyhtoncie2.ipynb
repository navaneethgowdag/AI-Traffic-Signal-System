```python
class NamespaceDemo:
    class_var = "I'm a class variable"

    def __init__(self):
        self.instance_var = "I'm an instance variable"

    def show_namespace(self):
        local_var = "I'm a local variable"
        print(f"Local namespace: {locals()}")
        print(f"Instance namespace: {self.__dict__}")
        print(f"Class namespace: {NamespaceDemo.__dict__}")


print("\nDemonstrating namespaces:")
ns_obj = NamespaceDemo()
ns_obj.show_namespace()

print("\n--- 1.5 TYPES OF METHODS ---")
print("1. Instance Methods: Work with instance variables (use 'self')")
print("2. Class Methods: Work with class variables (use 'cls')")
print("3. Static Methods: Don't access instance or class variables")
```

```python
class cal:
    pi = 3.14

    def __init__(self, name):
        self.name = name

    def add(self, a, b):
        print(f'adding {self.name}')
        print(a + b)

    @classmethod
    def mul(cls, a):
        print(cls.pi * a)

    @staticmethod
    def sub(a, b):
        print(a - b)

c = cal('nav')
c.add(3,4)
c.mul(9)
c.sub(8,3)
```

```python
#inheritence and polymorphism
#Inheritence

class animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species
    def display(self):
        print(f'{self.name} belongs to {self.species}')

class dog:
    def __init__(self,age):
        self.age = age
    def display1(self):
        animal.__init__(self,'Dog', 'goldef')
        print(f'{self.age} is {self.name} and belongs to {self.species}')

a = dog(23)
b = animal('cat', 'golden')
b.display()
a.display1()
```

```python
#MRO
class A():
    def some(self):
        print(1)

class B(A):
    def some(self):
        print(2)

class C(A):
    def some(self):
        print(3)

class D(B,C):
    pass

print(D.__mro__)
print(D.mro())
```

```python
# Detailed MRO Example
print("\n\nDetailed MRO Example:")


class X:
    def method(self):
        print("X method")


class Y:
    def method(self):
        print("Y method")


class Z:
    def method(self):
        print("Z method")


class A1(X, Y):
    pass


class B1(Y, Z):
    pass


class C1(A1, B1):
    pass


print("Class C1 inherits from A1 and B1")
print("A1 inherits from X and Y")
print("B1 inherits from Y and Z")
print(f"\nMRO of C1: {[cls.__name__ for cls in C1.mro()]}")

c1_obj = C1()
c1_obj.method()  # Which method will be called?
```

```python
# MRO with super()
print("\n\nMRO with super():")


class First:
    def __init__(self):
        print("First __init__")
        super().__init__()


class Second:
    def __init__(self):
        print("Second __init__")
        super().__init__()


class Third(First, Second):
    def __init__(self):
        print("Third __init__")
        super().__init__()


print(f"MRO: {[cls.__name__ for cls in Third.mro()]}")
print("\nCalling Third():")
third_obj = Third()
```

```python
print("\n--- 2.5 POLYMORPHISM ---")
print("Polymorphism means 'many forms' - same method name, different behavior")

# 1. Method Overriding (Runtime Polymorphism)
print("\n1. METHOD OVERRIDING:")


class Bird:
    def fly(self):
        print("Most birds can fly")


class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")


class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")


# Polymorphism in action
print("\nPolymorphism with different bird types:")
birds = [Sparrow(), Penguin(), Bird()]

for bird in birds:
    bird.fly()  # Same method name, different behavior
```

```python
# 2. Polymorphism with Functions
print("\n2. POLYMORPHISM WITH FUNCTIONS:")


def animal_sound(animal):
    """Function that works with any object having make_sound method"""
    animal.make_sound()


class Cat:
    def make_sound(self):
        print("Meow!")


class Cow:
    def make_sound(self):
        print("Moo!")


class Duck:
    def make_sound(self):
        print("Quack!")


print("Same function, different objects:")
animal_sound(Cat())
animal_sound(Cow())
animal_sound(Duck())
```

```python
# 2.6 DUCK TYPING PHILOSOPHY
# -----------------------------------------------------------------------------
print("'If it walks like a duck and quacks like a duck, it must be a duck'")
print("Python doesn't check object type, only if it has the required method/attribute")


class Dog2:
    def speak(self):
        return "Woof!"


class Cat2:
    def speak(self):
        return "Meow!"


class Robot:
    def speak(self):
        return "Beep boop!"


def make_it_speak(thing):
    """This function doesn't care about the object type"""
    print(thing.speak())


print("\nDuck typing - all have speak() method:")
make_it_speak(Dog2())
make_it_speak(Cat2())
make_it_speak(Robot())
```

```python
print("\n--- 2.7 OPERATOR OVERLOADING ---")
print("Operator overloading allows custom behavior for operators")


class Point:
    """Point class with operator overloading"""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        """String representation"""
        return f"Point({self.x}, {self.y})"

    def __add__(self, other):
        """Overloading + operator"""
        return Point(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        """Overloading - operator"""
        return Point(self.x - other.x, self.y - other.y)

    def __eq__(self, other):
        """Overloading == operator"""
        return self.x == other.x and self.y == other.y

    def __lt__(self, other):
        """Overloading < operator"""
        return (self.x ** 2 + self.y ** 2) < (other.x ** 2 + other.y ** 2)


print("\nOperator overloading with Point class:")
p1 = Point(3, 4)
p2 = Point(1, 2)

print(f"p1 = {p1}")
print(f"p2 = {p2}")
print(f"p1 + p2 = {p1 + p2}")
print(f"p1 - p2 = {p1 - p2}")
print(f"p1 == p2: {p1 == p2}")
print(f"p1 < p2: {p1 < p2}")
```

```python
class BankAccount:
    """Parent class"""

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: ${amount}. New balance: ${self.balance}")
        else:
            print("Insufficient funds!")

    def display(self):
        print(f"Account: {self.account_number}, Balance: ${self.balance}")


class SavingsAccount(BankAccount):
    """Child class with overridden methods"""

    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def withdraw(self, amount):
        """Overriding withdraw method with different logic"""
        if amount <= self.balance - 500:  # Minimum balance requirement
            self.balance -= amount
            print(f"Withdrawn: ${amount}. New balance: ${self.balance}")
        else:
            print("Cannot withdraw! Minimum balance of $500 required.")

    def display(self):
        """Overriding and extending parent method"""
        super().display()
        print(f"Interest Rate: {self.interest_rate}%")


print("\nMethod overriding example:")
regular_account = BankAccount("12345", 1000)
savings_account = SavingsAccount("67890", 1000, 3.5)

print("Regular Account:")
regular_account.withdraw(800)

print("\nSavings Account:")
savings_account.withdraw(800)  # Will fail due to minimum balance

print("\nDisplay methods:")
regular_account.display()
print()
savings_account.display()
```

```python
# ABSTRACT CLASSES AND INTERFACES
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass
        
    def display(self):
        print('car brand is', self.brand)

class car(Vehicle):
    def start_engine(self):
        print(f'{self.brand} car started')
    def stop_engine(self):
        print(f'{self.brand} stopped')

t = car("Honda")
t.start_engine()
t.stop_engine()
t.display()
```

```python
#INTERFACES IN PYTHOn

from abc import ABC, abstractmethod

# Interface
class Vehicle(ABC):

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass


# Implementing class
class Car(Vehicle):
    def start_engine(self):
        print("Car started")

    def stop_engine(self):
        print("Car stopped")


# Object creation
c = Car()
c.start_engine()
c.stop_engine()
#if class contains only abstract methods then it is interface right
```
